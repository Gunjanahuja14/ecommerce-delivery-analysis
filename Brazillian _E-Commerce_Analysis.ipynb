**STEP 1 : IMPORTING LIBRARIES**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**STEP 2 : LOADING THE DATASET**

```python
orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
```

**STEP 5: MERGING THE DATA**

```python
df = orders.merge(customers, on="customer_id", how="left")
df = df.merge(order_items, on="order_id" , how="left")
df = df.merge(payments, on="order_id", how="left")
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
df['order_id'].nunique()
```

**STEP 6: DATA CLEANING**

```python
df = df[df['order_status'] == 'delivered']

df = df.dropna(subset=[
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
])

df['payment_value'] = df['payment_value'].fillna(df['payment_value'].median())
df['payment_type'] = df['payment_type'].fillna(df['payment_type'].mode()[0])
```

```python
cols = [
    'order_purchase_timestamp',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in cols:
    df[col] = pd.to_datetime(df[col])
```

```python
print(df.isnull().sum())
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

**STEP 7: DATA TRANSFORMATION**

```python

df['delivery_time'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
df['delay'] = (df['order_delivered_customer_date'] - df['order_estimated_delivery_date']).dt.days
df['is_delayed'] = df['delay'] > 0


df['order_month'] = df['order_purchase_timestamp'].dt.month
df['order_day'] = df['order_purchase_timestamp'].dt.day_name()
df[['delivery_time','delay']].describe()
```

**STEP 8: UNIVARIANT ANALYSIS**

```python
sns.histplot(df['delivery_time'], bins=30)
plt.title("Distribution of Delivery Time")
plt.show()

sns.histplot(df['delay'], bins=30)
plt.title("Distribution of Delay")
plt.show()
```

**STEP 9: BIVARIATE ANALYSIS**

```python
df.groupby('order_day')['delay'].mean().sort_values()

sns.barplot(x='order_day', y='delay', data=df)
plt.title("Delay vs Day of Order")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('order_month')['delivery_time'].mean()
sns.lineplot(x='order_month', y='delivery_time', data=df)
plt.title("Delivery Time by Month")
plt.show()
```

```python
df.groupby('payment_type')['delay'].mean()
sns.boxplot(x='payment_type', y='delay', data=df)
plt.title("Delay by Payment Type")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='payment_value', y='delay', data=df)
plt.title("Price vs Delay")
plt.show()
```

**STEP 10 : CORRELATION ANALYSIS**

```python
df[['delivery_time','delay','payment_value']].corr()
sns.heatmap(df[['delivery_time','delay','payment_value']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

**STEP 11 : DATA VISUALISATION**

```python
sns.histplot(df['delivery_time'], bins=30, kde=True)
plt.title("Distribution of Delivery Time")
plt.xlabel("Days")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df['delay'], bins=30, kde=True)
plt.title("Distribution of Delivery Delay")
plt.show()
```

```python
df['is_delayed'].value_counts(normalize=True) * 100
sns.countplot(x='is_delayed', data=df)
plt.title("Delayed vs Non-Delayed Orders")
plt.show()
```
